```python
import numpy as np
```

```python
units_sold_2024 = np.array(
    [
        [5,8,3],
        [10,3,15],
        [15,20,4]
    ]
)

units_sold_2024
```

```python
units_sold_2025 = np.array(
    [
        [10,5,4],
        [25,1,40],
        [5,11,1]
    ]
)

units_sold_2025
```

<h2>Total Number of unit sold in 2024 and 2025</h2>

```python
units_sold_2024 + units_sold_2025
```

<h2>Sales unit growth from 2024 to 2025</h2>

```python
units_sold_2025 - units_sold_2024
```

<h2>Percentage Growth in Sales from 2024 to 2025</h2>

```python
((units_sold_2025 - units_sold_2024)/units_sold_2024)*100
```
