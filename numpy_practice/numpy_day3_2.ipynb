```python
import numpy as np
```

```python
units_sold_2024 = np.array(
    [
        [5,8,3],
        [10,3,15],
        [15,20,4]
    ]
)

units_sold_2024
```

```python
units_sold_2025 = np.array(
    [
        [10,5,4],
        [25,1,40],
        [5,11,1]
    ]
)

units_sold_2025
```

<h2>Total Number of unit sold in 2024 and 2025</h2>

```python
units_sold_2024 + units_sold_2025
```

<h2>Sales unit growth from 2024 to 2025</h2>

```python
units_sold_2025 - units_sold_2024
```

<h2>Percentage Growth in Sales from 2024 to 2025</h2>

```python
((units_sold_2025 - units_sold_2024)/units_sold_2024)*100
```

<h2>Prices of Each Product</h2>

```python
prices = np.array(
    [
        [160,100,85],
        [100,86,70],
        [60,50,55]
    ]
)

prices
```

<h2>Revenue of products in 2024</h2>

```python
revenue_2024 = units_sold_2024 * prices
revenue_2024
```

<h2>Discount 25% on prices</h2>

```python
discount = revenue_2024 * 0.25
discount
```

```python
net_revenue_2024 = revenue_2024 - discount

net_revenue_2024
```

<h1>Slicing Of Numpy Arrays</h1>

```python
arr = np.array(
[
    ["","India","China","USA"],
    ["Fold 4",5,10,3],
    ["Iphone 14",15,10,25],
    ["Oneplus 14",25,30,15]
])
arr
```

```python
arr
```
