# Numpy Space Efficient

<h1 style="background-color:red; color:yellow; padding:20px">Numpy Markdown</h1>

```python
import sys
```

```python
numbers = list(range(1,1001))
print(f"Numbers = {numbers[:10]}")
print(f"Type of Numbers = {type(numbers)}")
print(f"Last 10 Numbers = {numbers[:-10:-1]}")
print(f"Length of Numbers = {len(numbers)}")
print(f"Size of one element in list  = {sys.getsizeof(numbers[0])} bytes")
print(f"Total Size of list  = {sys.getsizeof(numbers[0])*len(numbers)} bytes")
```

# Numpy Arrays

```python
import numpy as np
```

```python
numpy_numbers = np.arange(1,1001, dtype="int16")
numpy_numbers[:5]
```

```python

print(f"Numpy Array = {numpy_numbers[:10]}")
print(f"Type of Numpy Array = {type(numpy_numbers)}")
print(f"Last 10 Elements of Numpy Array = {numpy_numbers[:-10:-1]}")
print(f"Length of Numpy Array = {len(numpy_numbers)}")
print(f"Size of one element in Numpy Array  = {numpy_numbers[0].nbytes} bytes")
print(f"Total Size of Numpy Array  = {numpy_numbers.nbytes} bytes")
```
