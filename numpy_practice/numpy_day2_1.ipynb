<h1 style="color:green">Time Efficient</h1>

```python
import numpy as np
import time as time
```

```python
list1 = list(range(1,100000000))
print(f"First 5 elements of list 1 = {list1[:5]}")
list2 = list(range(1,100000000))
print(f"First 5 elements of list 2 = {list2[:5]}")
```

```python
# Addition of both list and store in 3rd list
```

```python
start_time = time.time()
list3 =[]
# store time before starting loop
for i in range(0,len(list1)):
    list3.append(list1[i] + list2[i])

# store time after ending loop
end_time = time.time()
time_take = end_time - start_time
print(f"Time take to complete addition of list of size {len(list3)} = {time_take} seconds")
print(f"First 5 elements of list 3 = {list3[:5]}")
```

```python
# Time Efficint with numpy arrays
```

```python
arr1 = np.arange(1,100000000)
print(f"First 5 elements of arr1 1 = {arr1[:5]}")
arr2 = np.arange(1,100000000)
print(f"First 5 elements of arr1 1 = {arr2[:5]}")
```

```python
start_time = time.time()

arr3 = arr1+arr2

end_time = time.time()

time_taken = end_time - start_time

print(f"Time taken to complete addition of array of size {arr3.size} = {time_taken} seconds")
print(f"First 5 elements of arr3 3 = {arr3[:5]}")
```
