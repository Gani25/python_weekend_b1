<h1 style="color:red; text-align:center">Basic Operations Of Numpy</h1>

```python
import numpy as np
```

```python
arr = np.array([5,1,7,9,3])
print(arr)
print(arr.dtype)
print(f"Size of one element of numpy array is {arr[0].nbytes} bytes")
print(f"Total size of array = {arr.nbytes} bytes")
print(f"Total elements of array = {arr.size}")
print(f"Dimension of array {arr.ndim}")
print(f"Shape of array = {arr.shape}")
```

```python
arr = np.array(
    [
        [2,10,-6],
        [60,10,50],
        [100,250,600],
        [-12,-6,-5]
    ]
)
print(arr)
print(arr.dtype)
print(f"Size of one row of numpy array is {arr[0].nbytes} bytes")
print(f"Size of one element/column of numpy array is {arr[0,0].nbytes} bytes")
print(f"Total size of array = {arr.nbytes} bytes")
print(f"Total elements of array = {arr.size}")
print(f"Dimension of array {arr.ndim}")
print(f"Shape of array = {arr.shape}")
```
