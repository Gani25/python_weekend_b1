# Visualization With Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_sales = pd.DataFrame(
    {
        "Quarter":["Q1 2024","Q2 2024","Q3 2024","Q4 2024","Q1 2025","Q2 2025","Q3 2025","Q4 2025"],
        "Samsung Smart TV":[20,2,10,28,10,20,25, 31],
        "Refrigerator":[0,25,22,5,12,23,6,6],
        "Air Conditioner":[5,38,16,2,7,27,28, 3]
    }
)

df_sales
```

```python
df_sales.describe()
```

# Make a line plot of above data set

```python
plt.figure(figsize=(12,6))

plt.plot(df_sales["Quarter"], df_sales["Samsung Smart TV"], label="Samsung Smart TV", marker="o", color="grey"  )
plt.plot(df_sales["Quarter"], df_sales["Refrigerator"], label="Refrigerator", marker="p", color="#c4ae3f"  )
plt.plot(df_sales["Quarter"], df_sales["Air Conditioner"], label="Air Conditioner", marker="D", color="red" )

plt.legend()
plt.grid(linestyle="--", alpha=0.5)

plt.title("Unit of Product Sales in Quarters", fontsize=20, fontweight="bold", pad=35)

plt.xlabel("Quarters", fontsize=12)
plt.ylabel("Unit Sold (K)", fontsize=12)
custom_range = range(0,41,2)
plt.yticks(custom_range, fontsize=10)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```

## Scatter Plot

```python
plt.figure(figsize=(12,6))

plt.scatter(df_sales["Quarter"], df_sales["Samsung Smart TV"], label="Samsung Smart TV", color="grey"  )
plt.scatter(df_sales["Quarter"], df_sales["Refrigerator"], label="Refrigerator", color="#c4ae3f"  )
plt.scatter(df_sales["Quarter"], df_sales["Air Conditioner"], label="Air Conditioner", color="red", linewidth=1, edgecolors="black", s=100 )

plt.legend()
# plt.grid(linestyle="--", alpha=0.5)

plt.title("Unit of Product Sales in Quarters", fontsize=20, fontweight="bold", pad=35)

plt.xlabel("Quarters", fontsize=12)
plt.ylabel("Unit Sold (K)", fontsize=12)
custom_range = range(0,41,2)
plt.yticks(custom_range, fontsize=10)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Pie Plot

```python
# Calculate total sales in both year
total_sales = df_sales[["Samsung Smart TV","Refrigerator","Air Conditioner"]].sum()

total_sales
```

```python
total_sales.index
```

```python
plt.figure(figsize=(6,6))

plt.pie(
        total_sales, 
        labels=total_sales.index,
        startangle=0,
        colors=["lightgreen","orange","lightblue"],
        autopct="%.2f%%",
        wedgeprops={"edgecolor":"black"},
        shadow=True,
        explode=(0.2,0,0),
        labeldistance=0.3,
        pctdistance=1.2
        )
plt.title("Market Share Based on Total Products Sold",pad=30)
plt.show()
```
