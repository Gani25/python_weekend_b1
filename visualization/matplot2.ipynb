```python
import seaborn as sns
import numpy as np
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df_marks["Marks"], kde=True)
plt.xticks(range(0,101,5))

plt.xlabel(xlabel="Marks", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
sns.set_style("darkgrid")
hist = sns.histplot(df_marks["Marks"], binwidth=10, kde=True, color="skyblue", edgecolor="black", alpha=0.5)

for single_patch in hist.patches:
    height = single_patch.get_height()
    if height > 0:
        hist.text(
            single_patch.get_x() + single_patch.get_width()/2,
            height + 0.3,
            f"{height}",
            fontweight="bold",
            fontsize=10
                  
        )

plt.xticks(np.arange(0,101,5))
plt.xlabel("Marks Distribution", fontsize=14, labelpad=20, fontweight="bold")
plt.ylabel("Student Counts", fontsize=14, labelpad=20, fontweight="bold")
plt.title("Student Marks Distribution", fontsize=16, fontweight="bold", color="darkblue", pad=25)
plt.tight_layout()
plt.show()
```
