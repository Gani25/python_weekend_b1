# CRUD from Python With MySQL

```python
from sqlalchemy import create_engine, text
import pandas as pd
# import urllib.parse as url_parse
from urllib.parse import quote_plus
```

```python
# Only 1 db and username,pw,db is same
# If different then we need parameterized constructor

#("<dialect>+<driver>://<username>:<password>@<server_url:port>/<database_name>")
class CreateMySQLEngineInstance:

    def __init__(self):
        # password = quote_plus("abdul@123")
        password = "root"
        self.engine = create_engine(f"mysql+pymysql://root:{password}@localhost:3306/python_db")
    
    def get_engine(self):
        return self.engine
```

```python
# Test connection
mysql_instance = CreateMySQLEngineInstance()

engine = mysql_instance.get_engine()
print(engine)

query = "select * from course"

# Table in form of dataframe

df = pd.read_sql(query,engine)

df
```

```python
mysql_instance = CreateMySQLEngineInstance()

engine = mysql_instance.get_engine()

# Insert -> course_name, course_duration
course_name = input("Enter course name: ")
course_duration = input("Enter course duration: ")

insert_query = text("insert into course(name, duration) values (:cname, :cdur)")

# connection = engine.connect()

# connection.close()
with engine.connect() as connection:
    result = connection.execute(insert_query,{"cname":course_name,"cdur":course_duration})
    connection.commit()
    print(f"Insert successfully, Rows affected = {result.rowcount}")
```

```python
def show_course_table():
    mysql_instance = CreateMySQLEngineInstance()

    engine = mysql_instance.get_engine()


    query = "select * from course"

    # Table in form of dataframe

    df = pd.read_sql(query,engine)

    return df
```
