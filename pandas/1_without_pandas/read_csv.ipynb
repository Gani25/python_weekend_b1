<h2>Find Average Ratings, Minimum Rating, Maximum IMDB Rating of all movies</h2>

```python
movies_ratings = list()

for row in data:
    if(row[3] != "NULL"):
        movies_ratings.append(float(row[3]))

movies_ratings
```

```python
type(movies_ratings[0])
```

```python
average_rating = sum(movies_ratings) / len(movies_ratings)

max_rating = max(movies_ratings)
min_rating = min(movies_ratings)

print(f"~~~All Movies Rating Summary~~~")
print(f"Average IMDB Rating = {round(average_rating,2)}")
print(f"Maximum IMDB Rating = {max_rating}")
print(f"Minimum IMDB Rating = {min_rating}")
```


<h2>Find Average Ratings, Minimum Rating, Maximum IMDB Rating of Bollywood movies</h2>

```python
movies_ratings_bolly = list()

for row in data:
    if(row[3] != "NULL" and row[1] == "Bollywood"):
        movies_ratings_bolly.append(float(row[3]))

movies_ratings_bolly
```

```python
average_rating_bolly = sum(movies_ratings_bolly) / len(movies_ratings_bolly)

max_rating_bolly = max(movies_ratings_bolly)
min_rating_bolly = min(movies_ratings_bolly)

print(f"~~~Bollywood Movies Rating Summary~~~")
print(f"Average IMDB Rating = {round(average_rating_bolly,2)}")
print(f"Maximum IMDB Rating = {max_rating_bolly}")
print(f"Minimum IMDB Rating = {min_rating_bolly}")
```
