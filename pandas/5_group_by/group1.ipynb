# Group By Concept In Pandas

```python
import pandas as pd
```

```python
df_students = pd.DataFrame({
    "roll_no":[4,5,1,2,1,1,3,4,5,2,2,3,3],
    "branch":["Civil","Civil","Mech","Mech","Computer","Civil","Mech","Mech","Mech","Computer","Civil","Civil","Computer"],
    "name":["Rohit Sharma","Pranjal Gupta","Ashutosh Pandey","Rohini Sable","Mayuri Desai","Rohit Raj","Shubham Pandey","Anjali Sharma", "Salman Khan","John Wick","Joseph Carl","Atadipa Sharma","Shivam Pandey"],
    "percentage":[95,98,65,48,88,93,50,41,32,96,99,89,88]
})

df_students
```

```python
df_students.branch.value_counts()
```

```python
df_students.describe()
```

```python
df_mech = df_students[df_students["branch"]== "Mech"]
```

```python
df_mech
```

```python
# group by

df_group = df_students.groupby("branch")
```

```python
df_group
```

```python
df_group.get_group("Mech")
```

```python
df_group.get_group("Computer")
```

```python
# Display all groups in one go

for key, value in df_group:
    print("------------------------------------------------------")
    print(f"Branch = {key}")
    print(value)
```

```python
# Display all groups in one go

for key, value in df_group:
    print("------------------------------------------------------")
    print(f"Branch = {key}")
    print(value.reset_index())
```

```python
# Perform Basic Operations like min, max, average
for key, value in df_group:
    print("------------------------------------------------------")
    print(f"Branch = {key}")
    print(f"Maximum Percentage = {value.percentage.max()}")
    print(f"Minimum Percentage = {value.percentage.min()}")
    print(f"Average Percentage = {value.percentage.mean().round(2)}")
```
