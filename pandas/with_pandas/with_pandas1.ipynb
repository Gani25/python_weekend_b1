<h2>Average, Min and Max Rating Of All Movies</h2>

```python
print("~~~All Movies Rating Info~~~")
print(f"Average Rating = {dataframe["imdb_rating"].mean().round(2)}")
print(f"Max Rating = {dataframe["imdb_rating"].max()}")
print(f"Min Rating = {dataframe["imdb_rating"].min()}")
```
