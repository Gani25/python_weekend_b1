```python
import pandas as pd
```

```python
df_temp = pd.DataFrame({

    "city":["Mumbai","Pune","Delhi","Gurgaon","Bangaluru"],
    "temp":[28,32,24,22,15]
})
df_temp
```

```python
df_aqi = pd.DataFrame({

    "city":["Gurgaon","Hyderabad","Mumbai","Delhi","Bangaluru","Pune","Nashik"],
    "aqi":[205,170,155,605,355,190,112]
})
df_aqi
```

```python
# merge
pd.merge(df_temp,df_aqi,on="city")
```

```python
# Types of Join
pd.merge(df_aqi, df_temp,on="city")
```

```python
# Types of Join
pd.merge(df_aqi, df_temp,on="city", how="left")
```

```python
# Types of Join
pd.merge(df_aqi, df_temp,on="city", how="right")
```

```python
df_temp = pd.DataFrame({

    "city":["Mumbai","Pune","Delhi","Gurgaon","Bangaluru"],
    "temp":[28,32,24,22,15]
})
df_temp
```

```python
df_aqi = pd.DataFrame({

    "town":["Gurgaon","Hyderabad","Mumbai","Delhi","Bangaluru","Pune","Nashik"],
    "aqi":[205,170,155,605,355,190,112]
})
df_aqi
```

```python
# Types of Join
pd.merge(df_aqi, df_temp,left_on="town", right_on="city", how="left")
```

```python
# Types of Join
pd.merge(df_aqi, df_temp, how="cross")
```

```python
# Types of Join
pd.merge(df_aqi, df_temp, how="cross").shape
```
