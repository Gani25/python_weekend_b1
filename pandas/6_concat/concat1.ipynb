```python
df_us_weather = pd.DataFrame({
    "city":["New York","Chicago","Orlando"],
    "temperature":[28,25,16],
    "aqi":[10,15,7]
})
df_us_weather
```

```python
df_weather = pd.concat([df_india_weather,df_us_weather])

df_weather
```

```python
df_weather = pd.concat([df_india_weather,df_us_weather], ignore_index=True)

df_weather
```
